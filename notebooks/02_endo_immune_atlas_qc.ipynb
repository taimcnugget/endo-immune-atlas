```python
# -- Load libraries
import os
import scanpy as sc
from matplotlib import pyplot as plt
import seaborn as sns
from matplotlib.ticker import NullLocator
import pandas as pd
import numpy as np
```

```python
# -- Paths
input_path = "/content/drive/MyDrive/endo-immune-atlas/data/processed/GSE179640_raw.h5ad"
output_path_data = "/content/drive/MyDrive/endo-immune-atlas/data/processed"
output_path_figures = "/content/drive/MyDrive/endo-immune-atlas/results/preprocessing"

os.makedirs(output_path_figures, exist_ok=True)
```

```python
# -- Set palettes
patient_colors_list = [
    sns.husl_palette(n_colors=1, h=h, s=0.9, l=0.65).as_hex()[0]
    for h in np.linspace(0.01, 0.70, 14)
]

# map to patient IDs
patient_ids = ["C01", "C02", "C03",
               "E01", "E02", "E03", "E04", "E05",
               "E06", "E07", "E08", "E09", "E10", "E11"]

patient_palette = dict(zip(patient_ids, patient_colors_list))

# tissue types
tissue_palette = {
    "Ctrl": "#A3C5B7",
    "EuE": "#373F70",
    "EcP": "#596C94",
    "EcO": "#3B6341"
}
```

```python
combined = sc.read_h5ad(input_path)
combined.layers["counts"] = combined.X.copy()
```

```python
# --- Quality Control
# hemoglobin genes (hemos)
combined.var["hemos"] = combined.var_names.str.contains("^HB[^(P)]")

# ribosomal genes (ribos)
combined.var["ribos"] = combined.var_names.str.startswith(("RPS", "RPL"))

# mitochondrial genes (mitos)
combined.var["mt"] = combined.var_names.str.contains("MT-")
```

```python
sc.pp.calculate_qc_metrics(combined,
                           qc_vars = ["mt"],
                           percent_top = None,
                           inplace = True,
                           log1p= False)
```

```python
# -- Preprocessing figures

sc.pl.violin(combined,
    ["n_genes_by_counts"],
    groupby = "patient_id",
    jitter=0.4,
    rotation = 45,
    palette = patient_palette,
    ylabel = "Number of Genes by Counts",
    xlabel = "Patient_ID",
    show = False
)

plt.savefig(
    os.path.join(output_path_figures, "02_qc_n_genes_by_counts_by_patient.png"),
    bbox_inches='tight',
    dpi=300
)
plt.close()


sc.pl.violin(combined,
    ["total_counts"],
    groupby = "patient_id",
    jitter=0.4,
    rotation = 45,
    palette = patient_palette,
    ylabel = "Total Counts",
    xlabel = "Patient_ID",
    show = False
)
plt.savefig(
    os.path.join(output_path_figures, "02_qc_total_counts_by_patient.png"),
    bbox_inches='tight',
    dpi=300
)
plt.close()

sc.pl.violin(combined,
    ["pct_counts_mt"],
    groupby = "patient_id",
    jitter=0.4,
    rotation = 45,
    palette = patient_palette,
    ylabel = "% counts Mitochondrial",
    xlabel = "Patient_ID",
    show = False
)

plt.savefig(
    os.path.join(output_path_figures, "02_qc_pct_counts_mt_by_patient.png"),
    bbox_inches='tight',
    dpi=300
)
plt.close()
```

```python
combined.obs.groupby("patient_id")["pct_counts_mt"].describe()
```

```python
plt.scatter(
    combined.obs["n_genes_by_counts"],
    combined.obs["pct_counts_mt"],
    s=0.5,
    alpha=0.3
)
plt.xlabel("Number of Genes by Counts")
plt.ylabel("% counts Mitochondrial")
plt.axhline(y=25, color="red", linestyle="--")

plt.savefig(
    os.path.join(output_path_figures, "02_qc_mito_scatter.png"),
    bbox_inches='tight',
    dpi=300
)
plt.close()
```

```python
# -- Checking mt distribution to check on how to deal with QC parameters

samples = combined.obs["sample_id"].unique()
fig, axes = plt.subplots(4, 6, figsize=(20, 14))
axes = axes.flatten()

for i, sample in enumerate(sorted(samples)):
    mask = combined.obs["sample_id"] == sample
    axes[i].scatter(
        combined.obs.loc[mask, "n_genes_by_counts"],
        combined.obs.loc[mask, "pct_counts_mt"],
        s=0.5,
        alpha=0.3
    )
    axes[i].axhline(y=25, color="red", linestyle="--", linewidth=0.8)
    axes[i].set_title(sample, fontsize=8)
    axes[i].set_xlabel("Number of Genes", fontsize=7)
    axes[i].set_ylabel("% counts Mitochondrial", fontsize=7)

plt.tight_layout()

plt.savefig(
    os.path.join(output_path_figures, "02_qc_mito_scatter_per_sample_id.png"),
    bbox_inches='tight',
    dpi=300
)
plt.close()
```

```python
# -- Filter low quality cells
min_genes = 500
min_UMI = 1000
max_UMI = 100000

sc.pp.filter_cells(combined, min_genes=min_genes)
sc.pp.filter_cells(combined, min_counts=min_UMI)
sc.pp.filter_cells(combined, max_counts=max_UMI)
```

```python
# -- Remove cells with high mt %
mt_threshold = 25

combined.obs["outlier_mt"] = combined.obs["pct_counts_mt"] > mt_threshold

combined = combined[~combined.obs["outlier_mt"]].copy()

print(f"Cells remaining: {combined.n_obs}")
print(combined.obs.groupby("sample_id").size())
```

```python
# -- Doublet removal
sc.pp.scrublet(combined, batch_key="sample_id")
combined = combined[~combined.obs["predicted_doublet"]].copy()
```

```python
# -- Filter genes
min_cells = 3
sc.pp.filter_genes(combined, min_cells = min_cells)
print(f"Cells remaining: {combined.n_obs}")
```

```python
# -- Tan et. al figure 1

# B,1
fig, ax = plt.subplots(figsize=(6, 10))

sns.violinplot(
    data=combined.obs,
    x="total_counts",
    y="patient_id",
    palette=patient_palette,
    orient="h",
    ax=ax
)


ax.set_xscale("log")
ax.invert_xaxis()
ax.set_xlim(50000, 1000)
ax.set_xticks([50000, 10000, 1000])
ax.set_xticklabels(["50k", "10k", "1k"])

ax.set_xlabel("UMIs")
ax.set_ylabel("Patient ID")
sns.despine(left=True)

ax.xaxis.set_minor_locator(NullLocator())

plt.savefig(
    os.path.join(output_path_figures, "02_qc_umi_by_patient_id.png"),
    bbox_inches='tight',
    dpi=300
)
plt.close()
```
